```python
#Module for crossover operator

#-------------------------------------------------------------------------------------------------------------------------------

def crossover(list1, list2):
    temp1=list1[:]
    temp2=list2[:]
    
    out=True
    flag=0
    x=1
    
    while x==1:                              #Check if both list have mutually exclusive values else perform again for that
        if out==False and flag==1:
            break
            
        num=random.sample(range(42),2)
        maxr=num[0]
        minr=num[1]

        if maxr<minr:
            t=maxr
            maxr=minr
            minr=t
            
        t1=maxr-minr
        
        if t1>=4:
            flag=1

        l1=[]
        l2=[]

        c=0

        for i in range(minr, maxr+1):
            l1.append(list1[i])
            l2.append(list2[i])
            c+=1
        
        out = any(check in l1 for check in l2) 
        
        if out==True:
            flag=0
    
    #print(maxr)
    #print(minr)
    #print(l1)
    #print(l2)
    
    for i in range(0, 42): 
        for p in range(len(l2)):       #With c, error
            if list1[i]==l2[p]:
                list1[i]=l1[p]
                
    for i in range(0, 42):
        for p in range(len(l1)):
            if list2[i]==l1[p]:
                list2[i]=l2[p]

    p=0
                
    for i in range(minr, maxr+1):
        #for p in range(len(l2)):
        list1[i]=l2[p]
        p+=1
    
    p=0
            
    for i in range(minr, maxr+1):
        #for p in range(len(l1)):
        list2[i]=l1[p]
        p+=1
        
    #print(list1)
    #print(list2)
    #print(fitness(list1))
    #print(fitness(list2))
    #print(fitness(temp1))
    #print(fitness(temp2))
    
    if fitness(list1)<=fitness(temp1) and fitness(list2)<=fitness(temp2):
         return list1, list2
    elif fitness(list1)>fitness(temp1) and fitness(list2)>fitness(temp2):
         return temp1, temp2
    elif fitness(list1)<=fitness(temp1) and fitness(list2)>fitness(temp2):
         return list1, temp2
    elif fitness(list1)>fitness(temp1) and fitness(list2)<=fitness(temp2):
        return temp1, list2
            
print(lc1)
print(lc2)
crossover(lc1, lc2)
print(lc1)
print(lc2)
    
    
```

```python
#Module for mutation operator

#-------------------------------------------------------------------------------------------------------------------------------
def mutate(list):
    l1=list[:]
    print(l1)
    
    num=random.sample(range(42),2)
    maxr=num[0]
    minr=num[1]
    
    if maxr<minr:
        t=maxr
        maxr=minr
        minr=t
        
    print(maxr)
    print(minr)
    t=l1[maxr]
    l1[maxr]=l1[minr]
    l1[minr]=t
    return l1

print(lc1)
lcp=mutate(lc1)
print(lcp)
print(lc1)
```

```python
#Module for fitness operator

#-------------------------------------------------------------------------------------------------------------------------------

def fitness(list):
    #print(list)
    l1=[]
    
    for i in range(len(list)-1):
        l1.append(lc[list[i]][list[i+1]])
    
    arr=np.array(l1)
    #print(l1)
    #print(len(l1))
    #print(arr)
    
    sum=0
    
    for p in arr:
        sum+=p**2
    
    #print(sum)
    #print(np.sum(arr))
    sum=m.sqrt(sum)
    return np.sum(arr)
    
c1=fitness(lc1)
print(c1)
        
```

```python
#Module for probability operator

#-------------------------------------------------------------------------------------------------------------------------------

def probability(p, q):
    x=random.uniform(0, 10)
    n=(p/q)*10
    
    #print(n)
    #print(x)
    
    if x<n:
        return True
    else:
        return False

print(probability(6,8))
```

```python
#Module for emp

#-------------------------------------------------------------------------------------------------------------------------------

main = [[0]*42 for i in range(60)]

#list12[0]=random.sample(range(42), 42)

for i in range(0, 60):
        main[i]=random.sample(range(42),42)

#print(main)

global bestone
bestone=0

def great(ll):
    min=10000
    
    for l in ll:
        if min>fitness(l):
            min=fitness(l)
            bestone=ll.index(l)
    return min

print(great(main))

def emp():
    count=0
    for l in main:
        #print(count)
        count+=1
        #print(great(main))
        x=probability(4.5, 5)
        y=probability(4, 5)
        #print(x)
        #print(y)
        
        if x==True and y==True:
            #print(l)
            temp=reverse(l)
            if fitness(temp)<=fitness(l):
                #print(l)
                #print(fitness(l))
                l=temp
                #print(l)
                #print(fitness(l))
                main[count-1]=l
                #print(main[count-1])
                if main.index(l)<59:
                    l, main[count]=crossover(l, main[count])
            else: 
                if main.index(l)<59:
                    l, main[count]=crossover(l, main[count])                          #Assumption that crossover doesn't change fitness
        
        elif x==True and y==False:
            temp=reverse(l)
            if fitness(temp)<=fitness(l):
                l=temp
        
        elif x==False and y==True:
            
            c=0
            if main.index(l)<59:
                l, main[count]=crossover(l, main[count])
        
        elif x==False and y==False:
            continue

emp()
print(great(main))
#print(count)
print(main)
```

```python
#Module for onl

#-------------------------------------------------------------------------------------------------------------------------------

#print(main)

print(great(main))

p1=0

for l in main:
    p1+=1/(fitness(l))

print(p1)
print(main)

def onl():
    count=0
    for l in main:
        count+=1
        #print("llll")
        c=1/(fitness(l)*p1)
        #print(c)
        c1=probability(c, 1)
        
        if c1==False:
            continue
        
        x=probability(4.5, 5)
        y=probability(4, 5)
        
        if x==True and y==True:
            temp=reverse(l)
            if fitness(temp)<fitness(l):
                l=temp
                main[count-1]=l
                if main.index(l)<59:
                    l, main[count]=crossover(l, main[count])
            else: 
                if main.index(l)<59:
                    l, main[count]=crossover(l, main[count])                          #Assumption that crossover doesn't change fitness
        
        elif x==True and y==False:
            temp=reverse(l)
            if fitness(temp)<fitness(l):
                l=temp
        
        elif x==False and y==True:
            
            c=0
            if main.index(l)<59:
                l, main[count]=crossover(l, main[count])
        
        elif x==False and y==False:
            continue
            
onl()
print(great(main))

```

```python
#Module for scout

#-------------------------------------------------------------------------------------------------------------------------------

def exp(p):
    temp=mutate(main[p])
    
    if fitness(temp)<=fitness(main[p]):
        main[p]=temp
```
